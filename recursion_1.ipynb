# Depth first search on linked list 

```python
import inspect
import time
import matplotlib.pyplot as plt
```

```python
# Binary trees : https://www.geeksforgeeks.org/tutorial-on-binary-tree/
# Creating a node with pointing to two nodes containing the value in data 

class Node:
   def __init__(self, data=None, leftnode = None, rightnode = None):
      self.data = data
      self.left = leftnode
      self.right = rightnode

class BinaryTree:
   def __init__(self):
      self.head = None
```

```python
tree1 = BinaryTree()
tree1.head = Node("1")
e2 = Node("2")
e3 = Node("4")
e4 = Node("5")
e5 = Node("6")

# Link first Node to second node
tree1.head.left = e2
e2.left = e4

# Link second Node to third node
tree1.head.right = e3
e3.left = e5
```

```python
tree2 = BinaryTree()
tree2.head = Node("1")
e2 = Node("2")
e3 = Node("4")
e4 = Node("5")
e5 = Node("6")
e6 = Node("5")
e7 = Node("6")

# Link first Node to second node
tree2.head.left = e2
e2.left = e4
e4.left = e6

# Link second Node to third node
tree2.head.right = e3
e3.left = e5
e5.left = e7
```

```python
tree3 = BinaryTree()
tree3.head = Node("1")
e2 = Node("2")
e3 = Node("4")
e4 = Node("5")
e5 = Node("6")
e6 = Node("5")
e7 = Node("6")
e8 = Node("1")
e9 = Node("2")
e10 = Node("5")
e11 = Node("6")
e12 = Node("5")
e13 = Node("6")
e14 = Node("1")
e15 = Node("2")

# Link first Node to second node
tree3.head.left = e2
e2.left = e4
e4.left = e6
e6.left = e8
e8.left = e10
e10.left = e12
e12.left = e14

# Link second Node to third node
tree3.head.right = e3
e3.left = e5
e5.left = e7
e7.left = e9
e9.left = e11
e11.left = e13
e13.left = e15
```

```python
trees = [tree1, tree2, tree3]
```

## Recursion 

```python
def get_node(node):
    
    if node == None: # Return to the calling function if current nod eis None
        return
    
    num_calls_list.append(len(inspect.stack()))

    node_list.append(node.data) # Append to node_list if not None
    
    #print('left', node.data)
    a = get_node(node.left) # Traverse left node
    
    # print('right', node.data)
    b = get_node(node.right) # Traverse right node
    return 
```

```python
T = []
print_nodes_list = []
F_calls = [] # Total number of function calls
Max_stack_size = [] # Maximum stack space

for tree in trees:
    num_calls_list = []
    
    node_list = []
    node = tree.head

    start = time.time()
    get_node(node)
    end = time.time()
    
    T.append(end - start)
    
    print_nodes_list.append("-> ".join(node_list))
    
    F_calls.append(len(num_calls_list))
    Max_stack_size.append(max(num_calls_list))
```

## Iteration

```python
# Define a function to print the created linked list
def print_node(linked_list):
    node = linked_list.head
    stack, preorder_traversal = [], []
    while True:
        
        if node != None:
            preorder_traversal.append(node.data)
            stack.append(node)
            # print('Appending ', node.data, 'to stack ..', stack)
            node = node.left

        elif stack:
            node = stack.pop()
            node = node.right
        else:
            break
            
    return preorder_traversal #print("-> ".join(preorder_traversal))
```

```python
T = []
print_nodes_list = []

for tree in trees:
    node_list = []
    node = tree.head

    start = time.time()
    node_list = print_node(tree)
    end = time.time()
    
    T.append(end - start)
    
    print_nodes_list.append("-> ".join(node_list))
```
