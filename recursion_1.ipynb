## Fibonacci series 

```python
# Print the call stack python : # https://stackoverflow.com/questions/1156023/print-current-call-stack-from-a-method-in-code
# Measure pace taken up python call stack : https://pythonspeed.com/articles/measuring-memory-python/ |||| https://psutil.readthedocs.io/en/latest/#psutil.Process.memory_info
# System RAM usage : https://www.geeksforgeeks.org/psutil-module-in-python/#:~:text=1%29%20psutil.virtual_memory%20%28%29%20%E2%80%93%20This%20function%20gives%20system,of%20free%20physical%20memory%20this%20function%20is%20used.

# imports
import inspect
import time
import matplotlib.pyplot as plt
```

```python
def fibonacci(n):
    
    if n <= 1: return n
    
    '''for ind, i in enumerate(inspect.stack()):
       print(ind, i, '\n')'''
    
    num_calls_list.append(len(inspect.stack())) # Return a list of the frames pushed onto stack for executing the current line of code
    
    return fibonacci(n-1) + fibonacci(n-2)
```

```python
N = [2,3,5,8,10,13,15]
```

```python
A = [] # Fibonacci values
T = [] # Time taken
F_calls = [] # Total number of function calls
Max_stack_size = [] # Maximum stack space

for num in N:
    num_calls_list = []
    # total_available_memory = psutil.virtual_memory()[1]/(1024*1024*1024)
    start = time.time()
    ans = fibonacci(num)
    end = time.time()
    print('Fibonacci value of ', num, ':', ans)
    A.append(ans)
    T.append(end-start)
    F_calls.append(len(num_calls_list))
    num_calls_list_ = [ i - min(num_calls_list)+1 for i in num_calls_list]
    Max_stack_size.append(max(num_calls_list_))
    
    #print('Total RAM usage in Mb :',psutil.virtual_memory()[3]/(1024*1024*1024))
    # print('Total RAM used in Mb :', (total_available_memory - psutil.virtual_memory()[1]/(1024*1024*1024))*1024 )
    #print('Total time taken : ', str(time.time()- start))
```
